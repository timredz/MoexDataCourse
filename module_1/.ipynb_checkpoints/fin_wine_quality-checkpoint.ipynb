###  Построение модели

Задачу можно решать как регрессией, так и классификацией. Для разнообразия, возьмем регрессионную модель. Но нужно понимать, что в нашей задаче целевая переменная дискретная (7 различных значений), поэтому регрессионная модель, скорее всего выдаст не самый лучший результат. Но тем не менее, она тут применима.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

Попробуем пять различных алгоритма регресии: Lasso, ElasticNet, Gradient boosting, Random Forest и SVM. В качестве меры будем использовать mse(rmse) и еще дополнительно r2

```python
classifiers = [Lasso(),
               ElasticNet(),
               SVR(), 
               RandomForestRegressor(),
               GradientBoostingRegressor()] 
               
classifiers_name = ['Lasso',
                    'ElasticNet',
                    'SVR', 
                    'RandomForestRegressor',
                    'GradienBoostingRegressor']
```

```python
# Настройка параметров выбранных алгоритмов с помощью GridSearchCV 
n_folds = 5
scores = []
fits = []

lasso_params = {'alpha': (0.0001, 0.001, 1.0)}

elasticnet_params = {'alpha': (0.0001, 0.001, 0.1),
                    'l1_ratio': (0.1, 0.5)}

forest_params = {'n_estimators': [100, 300, 500],
                 'max_depth': list(range(7, 10)), 
                 'min_samples_leaf': list(range(10, 21, 10))}

svr_params = {'kernel' : ['rbf'], 
              'C': (.5, 1, 2)}

gbr_params = {'n_estimators': [100, 300, 500],
              'learning_rate':(0.1, 0.5, 1),
              'max_depth': list(range(3, 6)), 
              'min_samples_leaf': list(range(10, 31, 10))}

params = [lasso_params, elasticnet_params, svr_params, forest_params, gbr_params]
```

#### Кросс-валидация

```python
for i, each_classifier in enumerate(classifiers):
    clf = each_classifier
    clf_params = params[i]
    grid = GridSearchCV(clf, clf_params, 
                        cv=StratifiedKFold(y_train, n_folds=n_folds,
                        shuffle=False, random_state=42), 
                        n_jobs=-1, scoring="mean_squared_error")
    grid.fit(X_train, y_train)
    fits.append(grid.best_params_)
    clf_best_score = grid.best_score_
    scores.append(clf_best_score)
    print(classifiers_name[i], clf_best_score, "\n", grid.best_params_, "\n")
```

```python
%%time
clf_params = {'n_estimators': (250, 300, 350), 
              'learning_rate': (0.1, 0.2), 
              'min_samples_leaf': list(range(14, 29, 2)),
              'max_depth': list(range(4, 7))}

clf = classifiers[4]
grid = GridSearchCV(clf, clf_params, cv=n_folds, 
                    n_jobs=-1, scoring="mean_squared_error")
grid.fit(X_train, y_train)
clf_best_score = grid.best_score_
clf_best_params = grid.best_params_
clf_best = grid.best_estimator_
mean_validation_scores = []
print("Лучший результат", clf_best_score, 
      "лучшие параметры", clf_best_params)
```

### <center>Построение кривых валидации и обучения</center>

```python
def plot_with_std(x, data, **kwargs):
        mu, std = data.mean(1), data.std(1)
        lines = plt.plot(x, mu, '-', **kwargs)
        plt.fill_between(x, mu - std, mu + std, edgecolor='none',
                         facecolor=lines[0].get_color(), alpha=0.2)
        
def plot_learning_curve(clf, X, y, scoring, cv=5):
 
    train_sizes = np.linspace(0.05, 1, 20)
    n_train, val_train, val_test = learning_curve(clf,
                                                  X, y, train_sizes, cv=cv,
                                                  scoring=scoring)
    plot_with_std(n_train, val_train, label='training scores', c='green')
    plot_with_std(n_train, val_test, label='validation scores', c='red')
    plt.xlabel('Training Set Size'); plt.ylabel(scoring)
    plt.legend()

def plot_validation_curve(clf, X, y, cv_param_name, 
                          cv_param_values, scoring):

    val_train, val_test = validation_curve(clf, X, y, cv_param_name,
                                           cv_param_values, cv=5,
                                                  scoring=scoring)
    plot_with_std(cv_param_values, val_train, 
                  label='training scores', c='green')
    plot_with_std(cv_param_values, val_test, 
                  label='validation scores', c='red')
    plt.xlabel(cv_param_name); plt.ylabel(scoring)
    plt.legend()
```

### Часть 13. Выводы 

По результатам проведенного анализа, можно увидеть, что есть определенные зависимости и признаки показателя качества вина.
Выявлены определенные закаономерности - большое влияние на качество вина оказывает концентрация алкоголя. Чем ее больше, тем качественее вино. Считается, чем дольше выдержано вино, тем качественее. А при процессе выдержки увеличивается концентрация алкоголя. В наших данных этот факт мы увидели. 

Из наших данных видно, что вина высшего качества содержат больше сульфитов, а вина низкого качества содержат их мало. Дело в том, что в технологии винопроизводства используются сульфиты - диоксид серы, для приостанавливания нежелательного брожения (безопасный антисептик).

Также заметили, что у хороших вин низкая кислотность, у плохих она высокая. У вин есть особенность - скисать со временем, и если вино скисло - это уже не вино. И этот факт мы увидели в данных.

- Вина высокого качества характеризуются повышенной концентрацией алкоголя, низкой плотностью, повышенным уровнем pH и низкой концентрацией хлоридов. Имеют очень малый разброс по содержанию хлоридов и остаточного сахара.
- Вина низкого качества характеризуются низким уровенем диоксида серы (свободного) и остаточным сахаром, повышенным уровенем кислотности (изменчивая) и хлоридами.
